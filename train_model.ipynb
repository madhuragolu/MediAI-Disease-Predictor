```python
import joblib
```

```python
model = joblib.load("model.pkl")
```

```python
enco_pat = joblib.load("Disease_encoder.pkl") 
```

```python
trained_features = joblib.load("trained_features.pkl")
```

```python
pdf = pd.read_csv("dataset/symptom_precaution.csv")
```

```python
def predict_disease(sympt_txt):
    
    input_sympt = [sympt.strip() for sympt in sympt_txt.split(",")]
    features_dict = {feature: 0 for feature in trained_features}
    for sympt in input_sympt:
        if sympt in features_dict:
            features_dict[sympt] = 1
    input_df = pd.DataFrame([features_dict])
    prediction_encoded = model.predict(input_df)[0]
    predicted_disease = enco_pat.inverse_transform([prediction_encoded])[0]
    if predicted_disease in pdf['Disease'].values:
        row = pdf[pdf['Disease'] == predicted_disease].iloc[0]
        precautions_list = [
            row['Precaution_1'], row['Precaution_2'],
            row['Precaution_3'], row['Precaution_4']
        ]
    else:
        precautions_list = ["No precaution data available"] * 4
    print(f"\n Predicted Disease: {predicted_disease}\n")
    print(" Recommended Precautions:")
    for i, precaution in enumerate(precautions_list, start=1):
        print(f"{i}. {precaution}")


```
